```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Delhi_v2.csv')
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split

x=data.drop(['price'],axis=1)
y=data['price']
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2) #20% data reserved for evaluation
```

```python
train_data=x_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist()
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(numeric_only=True),annot=True , cmap='YlGnBu')
```

```python
data1=train_data.drop('Address',axis=1)

data1.info()

```

```python
data1
```

```python
data1=data1.join(pd.get_dummies(data1.Furnished_status,dtype=int)).drop(['Furnished_status'],axis=1)
data1=data1.join(pd.get_dummies(data1.neworold,dtype=int)).drop(['neworold'],axis=1)
data1=data1.join(pd.get_dummies(data1.type_of_building,dtype=int)).drop(['type_of_building'],axis=1)
data1=data1.drop('desc',axis=1)
```

```python
data1.info()
```

```python

data1.info()
```

```python
data1=data1.drop('Landmarks',axis=1)
data1=data1.drop('Status',axis=1)
data1=data1.drop('Unnamed: 0',axis=1)
```

```python
#model Evaluation
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train,y_train=data1.drop(['price'],axis=1),data1['price']
x_train_s = scaler.fit_transform(x_train)

reg = LinearRegression()
reg.fit(x_train_s,y_train)
```

```python
x_train.info()
```

```python
x_new = pd.DataFrame({
    'Price_sqft': [7000.0000],
    'Bedrooms': [3],
    'Bathrooms': [2],
    'parking':[1],
    'Furnished':[1],
    'Semi-Furnished':[0],
    'Unfurnished':[0],
    'Resale':[0],
    'Flat':[1], 
    'New Property':[1],
    'Individual House':[0], 
    'area':[1000],
    'Lift':[1],
    'latitude':[28.649000],
    'longitude':[77.369000],
    'Balcony': [3]
})
x_new = x_new[x_train.columns]


x_new_s = scaler.transform(x_new)

# 2. Predict the house price using the trained model
predicted_price = reg.predict(x_new_s)

# 3. Output the predicted price
print(f'Predicted House Price: INR {predicted_price[0]:,.2f}')

```

```python
test_data=x_test.join(y_test)
test_data=test_data.drop('Address',axis=1)
test_data=test_data.join(pd.get_dummies(test_data.Furnished_status,dtype=int)).drop(['Furnished_status'],axis=1)
test_data=test_data.join(pd.get_dummies(test_data.neworold,dtype=int)).drop(['neworold'],axis=1)
test_data=test_data.join(pd.get_dummies(test_data.type_of_building,dtype=int)).drop(['type_of_building'],axis=1)
test_data=test_data.drop('desc',axis=1)
test_data=test_data.drop('Landmarks',axis=1)
test_data=test_data.drop('Status',axis=1)
```

```python
x_test,y_test=test_data.drop(['price'],axis=1),test_data['price']
```

```python
x_test=x_test.drop('Unnamed: 0',axis=1)
```

```python
x_test.info()
```

```python
x_test_s=scaler.transform(x_test)
```

```python
reg.score(x_test_s,y_test)
```

```python
#random Forest regressor
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor()
forest.fit(x_train,y_train)
```

```python
forest.score(x_test,y_test)
```

```python
#Hyperparameter Tuning for better optimal solution
from sklearn.model_selection import GridSearchCV
forest = RandomForestRegressor()
param_grid ={
    "n_estimators":[100,200,300],
    "min_samples_split":[2,4],
    "max_depth": [None,4,8]
}
grid_search = GridSearchCV(forest,param_grid,cv=5,scoring="neg_mean_squared_error",return_train_score=True)
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_estimator_.score(x_test,y_test)
#project completed - model evaluation completed
```

```python
import joblib

# Save the best model found by GridSearchCV
best_model = grid_search.best_estimator_
joblib.dump(best_model, 'best_rf_model.pkl')
```
